```python
df_adv=pd.DataFrame({"TV":[1,2,3,4,12,15,15,14,16,14,14],"RADIO":[9,10,11,12,23,24,26,15,16,18,17],"NEWSPAPER":[11,12,13,14,24,32,36,34,52,62,52],"SALES":[90,100,110,120,150,142,350,250,203,250,620]})
```

```python
df_adv
```

```python
X=df_adv[["TV","RADIO","NEWSPAPER"]]
y=df_adv["SALES"]
```

```python
import statsmodels.api as sm
```

```python
X
```

```python
X=sm.add_constant(X)
```

```python
X
```

```python
model= sm.OLS(y,X).fit()
```
