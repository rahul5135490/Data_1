```python
# minimum share price that has gone for those specific year
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
# rule A shows year end specific
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
df_tesla.resample(rule='A').max()['Close'].plot()
```

```python
df_tesla.resample(rule='BQS').max()['Open'].plot()
```

```python
df_tesla.resample(rule='A').mean().plot(kind="bar")
```

```python
df_tesla['Open'].resample(rule='A').mean()
```

```python
df_tesla['Open'].resample(rule='M').mean()
```

```python
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar',figsize=(16,6))
```
